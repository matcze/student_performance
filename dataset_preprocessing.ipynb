```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
```

### 1. Loading dataset

The data used for this project come from: https://www.kaggle.com/datasets/grandmaster07/student-exam-performance-dataset-analysis

```python
dataset = pd.read_csv("dataset.csv")
```

### 2. Data Cleaning 

2.1 Check and remove duplaciates

```python
dataset.duplicated().sum() 
```

There are no duplicates in the dataset to remove.

2.2 Detect missing values

```python
dataset.isnull().sum()
```

There are three features with missing values: Teacher_Quality, Parental_Education_Level and Distance_from_Home. 

Since all three features are ordinal categorical variables, we will initially impute missing values using the mode.

```python
dataset["Teacher_Quality"].fillna(dataset["Teacher_Quality"].mode()[0], inplace = True)
dataset["Parental_Education_Level"].fillna(dataset["Parental_Education_Level"].mode()[0], inplace = True)
dataset["Distance_from_Home"].fillna(dataset["Distance_from_Home"].mode()[0], inplace = True)
```

```python
dataset.isnull().sum()
```

### 3. Data Preprocessing

**3.1 Binary Encoding**

There are four binary features: Extracurricular_Activities, Internet_Access, Learning_Disabilities, Gender.

```python
dataset["Extracurricular_Activities"].value_counts()
```

```python
dataset["Internet_Access"].value_counts()
```

```python
dataset["Learning_Disabilities"].value_counts()
```

```python
dataset["Gender"].value_counts()
```

```python
dataset["Gender"] = dataset["Gender"].replace({"Male": 1, "Female": 0})
dataset["Extracurricular_Activities"] = dataset["Extracurricular_Activities"].replace({"Yes": 1, "No": 0})
dataset["Internet_Access"] = dataset["Internet_Access"].replace({"Yes": 1, "No": 0})
dataset["Learning_Disabilities"] = dataset["Learning_Disabilities"].replace({"Yes": 1, "No": 0})
dataset["School_Type"] = dataset["School_Type"].replace({"Public": 1, "Private": 0})
```

```python
dataset.head()
```

**3.2 Ordinal encoding**

There are eight ordinals features: Parental_Involvement, Access_to_Resources, Motivation_Level, Family_Income, Teacher_Quality, Peer_Influence, Parental_Education_Level, Distance_from_Home

```python
ordinal_features = ["Parental_Involvement", "Access_to_Resources", "Motivation_Level", "Family_Income", "Teacher_Quality", "Peer_Influence", "Parental_Education_Level", "Distance_from_Home"]
```

```python
dataset["Distance_from_Home"].value_counts()
```

```python
categories = [
    ["Low", "Medium", "High"],      # Parental Involvment 
    ["Low", "Medium", "High"],        # Access to Resources
    ["Low", "Medium", "High"],        # Motivation Level 
    ["Low", "Medium", "High"],        # Family Income
    ["Low", "Medium", "High"],        # Teacher Quality
    ["Positive", "Neutral", "Negative"],        # Peer Influence
    ["High School", "College", "Postgraduate"],        # Parental Education Level
    ["Near", "Moderate", "Far"]      # Distance from Home
]
```

```python
ordinal_encoder = OrdinalEncoder(categories=categories)
```

```python
dataset[ordinal_features] = ordinal_encoder.fit_transform(dataset[ordinal_features])
```
